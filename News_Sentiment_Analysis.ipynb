#### Pull down actual article content

```python
# import libraries and use them to pull the articles content from the web
import requests
from bs4 import BeautifulSoup

# Create a function to pull the article content from the web
def get_article_content(url):
    # Make a GET request to fetch the raw HTML content
    html_content = requests.get(url).content

    # Parse the html content
    soup = BeautifulSoup(html_content, "html")

    # get the body element 
    body = soup.find("body")
    if body is None:
        return ""

    # we don't know what the article content is going to be, so we will use the find_all method to get all the elements
    article_text = ""
    for element in body.find_all("p"):
        article_text += "\n" + element.text

    return article_text


# Loop through each article URL and get the content
for url in article_df["article_url"]:
    # fill in the article text column with the article content
    article_df.loc[article_df["article_url"] == url, "article_text"] = get_article_content(url)

article_df.head()
```

### Send to Open AI for sentiment analysis

```python
# import libraries and use openai chat complete to generate a sentiment score for the article description
import openai

# Replace with your API key
openai.api_key = ""

def generate_sentiment(content, prompt, ticker):
    message = "Ticker: " + ticker + "\n" + "Content: " + content + "\n" + "Sentiment: "

    completion = openai.ChatCompletion.create(
        model="gpt-3.5-turbo",
        messages=[
            {"role": "system", "content": prompt},
            {"role": "user", "content": message}
        ]
    )
    print(completion.choices[0].message)
    return completion.choices[0].message["content"]


# Define the prompt
prompt = """You are a stock market agent with the goal of scoring the sentiment towards a stock given a ticker and an article. The sentiment score should be from 0 to 100. 100 represents the best possible sentiment.
Your output must be a JSON object with the following format: { Score: 100 }
"""

for description in article_df["article_description"]:
    # fill in the description sentiment column with the sentiment score
    article_df.loc[article_df["article_description"] == description, "description_sentiment"] = generate_sentiment(description, prompt, "AAPL")

article_df.head()
```

```python
# clean the description sentiment column
# { Score: 0.49 } -> 0.49
article_df["description_sentiment"] = article_df["description_sentiment"].str.replace(" Score: ", "")
article_df["description_sentiment"] = article_df["description_sentiment"].str.replace(" }", "")
# remove any non numeric characters
article_df["description_sentiment"] = article_df["description_sentiment"].str.replace("[^0-9.]", "")
# # remove any numeric characters after the third decimal place
article_df["description_sentiment"] = article_df["description_sentiment"].str.replace("\.[0-9]{3,}", "")
# keep the first two characters
article_df["description_sentiment"] = article_df["description_sentiment"].str[:2]



article_df.head()
```
